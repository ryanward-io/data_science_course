# Fundamental image classification

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn import metrics
from sklearn.model_selection import train_test_split

from sklearn import ensemble

dat = datasets.load_digits()

digits = pd.DataFrame(dat['data'])

digits['digit'] = dat['target']
digits
```

```python
dat['images'][-1]
```

```python
plt.imshow(dat['images'][-1], cmap='gray');
```

```python
y = digits['digit']
X= digits.drop('digit', axis=1)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clr = ensemble.RandomForestClassifier(n_estimators=1000, random_state=1)
```

```python
clr.fit(X_train, y_train)
```

```python
prediction = clr.predict(X_test)
```

```python
evaluate = pd.DataFrame({
    'actual': y_test,
    'predicted': prediction
})
```

```python
sns.scatterplot(data=evaluate, x='actual', y='predicted')
```

```python
metrics.accuracy_score(evaluate['actual'], evaluate['predicted'])
```

```python
metrics.confusion_matrix(evaluate['actual'], evaluate['predicted'])
```
